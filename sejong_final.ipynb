##### 건폐율 : 대지에 대한 건축면적의 비율입니다.
##### 건폐율 = (건축면적/대지면적)*100

##### 용적률 : 대지면적에 대한 건축물의 연면적의 비율
##### 용적률 = (연면적/대지면적)*100

##### 계약면적 = 기타공용면적(노인정,관리사무소) + 공급면적(전용면적+주거공용면적)

```python
# 시군구 추출
def preprocess_dong(data):
    if '시군구' in data.columns:
        data['읍면동'] = data['시군구'].str[8:11]
        return data
    elif '대지위치' in data.columns:
        data['읍면동'] = data['대지위치'].str[9:13]
        return data
    
# 연도,월 추출, 최근거래기간
def preprocess_year(data):
    #data[data['건축년도'].isna() == True]
    #data = data.dropna(axis=0)
    data['건축년도'] = data['건축년도'].astype(int).convert_dtypes()
    data['계약년'] = data['계약년월'].astype(str).str[:4]
    data['계약월'] = data['계약년월'].astype(str).str[4:6]
    data['최근거래기간'] = data['계약년'].astype(int) - data['건축년도'].astype(int)
    return data

# 집계함수 그룹 데이터프레임 생성
def preprocess_agggroup(data,col1,col2,func):
    data[col2] = data[col2].astype(float).convert_dtypes()
    pre_data = data[[col1,col2]].groupby([col1]).agg(func).reset_index().sort_values(by=col2,ascending=False)
    return pre_data

"""
# 최적의 단지 찾기
def preprocess_good(data,dx1,dy1,dz1):
    #pre_data = data[['읍면동',dx1,dy1]].groupby(['읍면동']).mean().reset_index()
    for col in data['읍면동'].unique():
        go_data = data[data['읍면동']==col]
        pre_data = data[['읍면동',dx1,dy1]].groupby(['읍면동']).mean().reset_index()
        idx = pre_data[pre_data['읍면동']==col].reset_index()['index'][0]
        go_data_count = go_data[(go_data[dx1] > pre_data[pre_data['읍면동']==col][dx1][idx]) & (go_data[dy1] < pre_data[pre_data['읍면동']==col][dy1][idx]) & (go_df3[dz1]>2013)]['읍면동'].count()
        print(col+'의 최적의 단지 개수 :',go_data_count)
        print('\n')

# 구룹별 x y 비교  시각화
def preprocess_betweengroup(data,group,xd,yd):
    pre_data = data[group].value_counts().reset_index()
    pre_data.columns = ['읍면동','거래횟수']
    pre_data_head = pre_data.sort_values(by='거래횟수',ascending=False).head(5)
    for head in pre_data_head['읍면동']:
        data_group = data[data['읍면동']==head]
        sns.lineplot(x=xd,y=yd,data=data_group) # 신뢰구간 기본값, ci = 'sd', 분포 ci =None 신뢰구간 지우기
"""

def preprocess_pyeong(data):
    data['전용면적에 따른 평수'] = data['전용면적(㎡)']*0.3025
    data['전용면적에 따른 평수'] = round(data['전용면적에 따른 평수']).astype(float).convert_dtypes()
    data['거래금액(만원)'] = data['거래금액(만원)'].astype(str).str.replace(',','')
    data['거래금액(만원)'] = data['거래금액(만원)'].astype(int)
    data['평당가격(만원)'] = data['거래금액(만원)']/data['전용면적에 따른 평수']
    return data
```
